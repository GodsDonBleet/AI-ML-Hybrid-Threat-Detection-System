# **Preprocessing - Encoding and Normalization**

```python
def encoding_map_maker(column):
  unique_values = column.unique()
  encoding_map = {}
  for i in range(len(unique_values)):
    encoding_map[unique_values[i]] = i + 1
  return encoding_map

non_num_columns = ['protocol_type', 'service', 'flag']
for col in x.columns[1:]:
  if col in non_num_columns:
    encoding_map = encoding_map_maker(x[col])
    x[col] = x[col].map(encoding_map)

x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(x)
df_scaled = pd.DataFrame(df_scaled, columns=x.columns)
df_scaled.head()
```

# **Preprocessing - PCA**

```python
#libs for PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#pca = PCA()
pca = PCA(n_components = 26)
df_pca = pca.fit_transform(df_scaled)
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
# Cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Plot the cumulative explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(
    range(1, len(cumulative_variance_ratio) + 1),
    cumulative_variance_ratio,
    marker='o',
    linestyle='--'
)
plt.title('Cumulative Explained Variance by Number of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()

# Mark the number of components explaining 95% of the variance
threshold = 0.95
n_components_95 = np.argmax(cumulative_variance_ratio >= threshold) + 1
plt.axhline(y=threshold, color='r', linestyle='--', label=f'95% Variance Threshold')
plt.axvline(x=n_components_95, color='r', linestyle='--', label=f'{n_components_95} Components')

plt.legend()
plt.show()

print(f"Number of components to explain 95% of the variance: {n_components_95}")
```
